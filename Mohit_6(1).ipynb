```python
results = list(apriori(data, min_support=0.01, min_confidence=0.4, min_lift=1.0))
```

```python
rules = []
supports = []
confidences = []

```

```python
# Extract rules, support, and confidence from the results
for result in results:
    support = result.support
    for rule in result.ordered_statistics:
        antecedent = tuple(rule.items_base)
        consequent = tuple(rule.items_add)
        confidence = rule.confidence
        rules.append(f"{antecedent} -> {consequent}")
        supports.append(support)
        confidences.append(confidence)

# Create a DataFrame from the extracted data
df_results = pd.DataFrame({
    'Rule': rules,
    'Support': supports,
    'Confidence': confidences
})

# Display the DataFrame
df_results
```
